# What is Numpy ? 
* Numpy Stands for numerical python.
* NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices

# Core Numpy Features
1. supports fast and efficient multidimensional arrays 
2. Excutes computations and mathematical calculation 
3. Performs linear operations 
4. Random number genaration 
5. Efficient way of storing and manipulating data 
6. Add,remove, update 
7. Only single type which is homogeneous 
8. Support wide varity of mathematical functions
