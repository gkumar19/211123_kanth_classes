```python
#Merge
df1 = pd.DataFrame({'type': ['fruit', 'fruit', 'vegetable'],
                  'price': [10, 12, 100],
                  'name1': ['mango', 'orange', 'potato']})

df2 = pd.DataFrame({'available': ['yes', 'no'],
                  'name2': ['mango', 'orange']})
```

```python
pd.concat([df1, df2], axis=1)
```

```python
pd.merge(df1.rename(columns={'name1': 'name'}),
         df2.rename(columns={'name2': 'name'}), on='name', how='outer')
```
